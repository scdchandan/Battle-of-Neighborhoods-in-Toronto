#### Data Extraction and Dataframe Creation

Step1: Importing Required Libraries

```python
import requests  # this module helps us to download a web page
import pandas as pd
```

Step2: Requesting Data from URL via GET method. A 200 response signifes successful connection and a successful HTTP request.

```python
url = 'https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'
data  = requests.get(url)
data
```

Step3: Using Pandas read_html to scrap data from the data received.

```python
toronto_data = pd.read_html(url, flavor='bs4')
```

The table data extracted is as follows. It is a list of tables and we will be using the first table only.

```python
toronto_data
```

The first table stores the required data. Each cell stores a string which contains teh postal code, borough name and list of neighborhoods separated by slashes.

```python
toronto_data[0]
```

```python
toronto_data = toronto_data[0]
toronto_data.shape
```

Step4: For each cell in the required table, we extract the substrings for postal code and borough if the borough is assigned. The slashes are replaced with commas to extract the neighborhoods as desired. An array of JSONs is created.

```python
table_values = []
for i in range(0,20):
    for j in range(0,9):
        cell = {}
        val = toronto_data.iloc[i,j]
        
        if(val[3:] != "Not assigned"):
            cell['PostalCode'] = val[:3]
            cell['Borough'] = val[3:].split('(')[0]
            cell['Neighborhood'] = val[3:].split('(')[1].split(')')[0].replace(' /',',').strip(" ")
#             print(cell)
            table_values.append(cell)

table_values
```

Step5: The array of JSONs created, is then converted into a Pandas Dataframe.

```python
toronto_df = pd.DataFrame(table_values)
toronto_df['Borough']=toronto_df['Borough'].replace({'Downtown TorontoStn A PO Boxes25 The Esplanade':'Downtown Toronto Stn A',
                                             'East TorontoBusiness reply mail Processing Centre969 Eastern':'East Toronto Business',
                                             'EtobicokeNorthwest':'Etobicoke Northwest','East YorkEast Toronto':'East York/East Toronto',
                                             'MississaugaCanada Post Gateway Processing Centre':'Mississauga'})
```

Step5: The following fucntion will be used to extract the venue list for each neighborhood based on latitude and longitude values of Postal Codes.

```python
def getNearbyVenues(names, latitudes, longitudes, radius=500):
    
    venues_list=[]
    for name, lat, lng in zip(names, latitudes, longitudes):
        print(name)
            
        # create the API request URL
        url = 'https://api.foursquare.com/v2/venues/explore?&client_id={}&client_secret={}&v={}&ll={},{}&radius={}'.format(
            CLIENT_ID, 
            CLIENT_SECRET, 
            VERSION, 
            lat, 
            lng, 
            radius
            )
            
        # make the GET request
        results = requests.get(url).json()["response"]['groups'][0]['items']
        
        # return only relevant information for each nearby venue
        venues_list.append([(
            name, 
            lat, 
            lng, 
            v['venue']['name'], 
            v['venue']['categories'][0]['name']) for v in results])

    nearby_venues = pd.DataFrame([item for venue_list in venues_list for item in venue_list])
    nearby_venues.columns = ['Neighborhood', 
                  'Neighborhood Latitude', 
                  'Neighborhood Longitude', 
                  'Venue', 
                  'Venue Category']
    
    return(nearby_venues)
```

Step6: The following values stores all nearby venues for each Neighborhood

```python
venues_in_toronto = getNearbyVenues(toronto_df_final['Neighborhood'], toronto_df_final['Latitude'], toronto_df_final['Longitude'])
```

Step7: ONE HOT encoding of the venue categories.

```python
toronto_venue_cat_one_hot = pd.get_dummies(venues_in_toronto[['Venue Category']], prefix="", prefix_sep="")
toronto_venue_cat_one_hot
```

Step8:  The dummies creadted already had a nieghborhood column which we are dropping here as it was labeled 1 only for 4 cases. 
        It may be utilized by a different user by changing its column name as we will add a column named Neighborhod to the             initial dataframe.

```python
toronto_venue_cat_one_hot = toronto_venue_cat_one_hot.drop(['Neighborhood'], axis = 1)
```

Step9: Neighborhood column is then added and is next moved to the first column of teh database.

```python
toronto_venue_cat_one_hot['Neighborhood'] = venues_in_toronto['Neighborhood'] 

# moving neighborhood column to the first column
fixed_columns = [toronto_venue_cat_one_hot.columns[-1]] + list(toronto_venue_cat_one_hot.columns[:-1])
toronto_venue_cat_one_hot = toronto_venue_cat_one_hot[fixed_columns]

toronto_venue_cat_one_hot.head()
```

```python
toronto_grouped = toronto_venue_cat_one_hot.groupby('Neighborhood').mean().reset_index()
toronto_grouped.head()
```

This function returns teh top N venues queried by the user. 

```python
def return_most_common_venues(row, num_top_venues):
    row_categories = row.iloc[1:]
    row_categories_sorted = row_categories.sort_values(ascending=False)
    
    return row_categories_sorted.index.values[0:num_top_venues]
```

Step10: We will be storing the top 20 venues.

```python
num_top_venues = 20

indicators = ['st', 'nd', 'rd']

# create columns according to number of top venues
columns = ['Neighborhood']
for ind in range(num_top_venues):
    try:
        columns.append('{}{} Most Common Venue'.format(ind+1, indicators[ind]))
    except:
        columns.append('{}th Most Common Venue'.format(ind+1))

# create a new dataframe
neighborhoods_venues_sorted = pd.DataFrame(columns=columns)
neighborhoods_venues_sorted['Neighborhood'] = toronto_grouped['Neighborhood']

for ind in range(toronto_grouped.shape[0]):
    neighborhoods_venues_sorted.iloc[ind, 1:] = return_most_common_venues(toronto_grouped.iloc[ind, :], num_top_venues)

neighborhoods_venues_sorted.head()
```

Step11: Running K means Algorithm with 5 clusters.

```python
# set number of clusters
k_num_clusters = 5

toronto_grouped_clustering = toronto_grouped.drop('Neighborhood', 1)

# run k-means clustering
kmeans = KMeans(n_clusters=k_num_clusters, random_state=0).fit(toronto_grouped_clustering)
kmeans
```

```python
kmeans.labels_[0:100]
```

Step12: Inserting the Cluster Indices for each Neighborhood

```python
neighborhoods_venues_sorted.insert(0, 'Cluster Labels', kmeans.labels_)
neighborhoods_venues_sorted
```

Step13: Adding in the Latitude and Longitude columns to facilitate Graph plotting

```python
toronto_merged = toronto_df_final

toronto_merged = toronto_merged.join(neighborhoods_venues_sorted.set_index('Neighborhood'), on='Neighborhood')

toronto_merged.head()
```

```python
toronto_merged_nonan = toronto_merged.dropna(subset=['Cluster Labels'])
```

```python
import matplotlib.cm as cm
import matplotlib.colors as colors
import numpy as np
```

Step14: Finally Plotting the Map of Toronto and the clustered neighborhoods.

```python
map_clusters = folium.Map(location=[latitude, longitude], zoom_start=11)

# set color scheme for the clusters
x = np.arange(k_num_clusters)
ys = [i + x + (i*x)**2 for i in range(k_num_clusters)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
for lat, lon, poi, cluster in zip(toronto_merged_nonan['Latitude'], toronto_merged_nonan['Longitude'], toronto_merged_nonan['Neighborhood'], toronto_merged_nonan['Cluster Labels']):
    label = folium.Popup('Cluster ' + str(int(cluster) +1) + '\n' + str(poi) , parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[int(cluster-1)],
        fill=True,
        fill_color=rainbow[int(cluster-1)]
        ).add_to(map_clusters)
        
map_clusters
```
